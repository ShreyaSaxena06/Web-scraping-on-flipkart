```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=95d38ec6-619e-4960-8a72-fa5902549b9e")
page.content
soup=BeautifulSoup(page.content,"html.parser")
print(soup.prettify)
```

```python
mobiles=soup.find_all("div",class_="_1AtVbE col-12-12")
print(len(mobiles))
```

```python
names=[]
prices=[]
front_camera=[]
for mobile in mobiles:
  name=mobile.find("div",class_="_4rR01T")
  names.append(name)
  price=mobile.find("div",class_="_30jeq3 _1_WHN1")
  prices.append(price)
  ul=soup.find("ul",class_="_1xgFaf")
  camera=ul.find_all("li")[2].text
  fc=camera.split(" | ")[1].split(" ")[0]
  front_camera.append(fc)
  
```

```python
for mobile in mobiles:
  ul=soup.find("ul",class_="_1xgFaf")
  print(ul.text)
```

```python
import pandas as pd
dic={"names":names,"prices":prices,"front_camera":front_camera}
df=pd.DataFrame(dic)
df.head()
df.to_csv("mobile_info.csv")
```
